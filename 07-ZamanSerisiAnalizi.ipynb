```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("datasets/AirPassengers.csv")
```

```python
df.head(8)
```

```python
df=df["#Passengers"]
```

```python
df=df.values
```

```python
df=df.reshape(-1,1)
df[0:8]
```

```python
df.shape
```

```python
scaler=MinMaxScaler()

df=scaler.fit_transform(df)
```

```python
df[0:8]
```

```python
def veri_olustur(data, time_step=1):
    X, Y = [], []

    for i in range(len(data) - time_step - 1):
        a=data[i:(i+time_step),0]
        X.append(a)
        Y.append(data[i+time_step,0])

    return np.array(X), np.array(Y)
```

```python
time_step=3
```

```python
train=df[0:100,:]
test=df[100:,:]
```

```python
x_train,y_train=veri_olustur(train,time_step)
```

```python
x_test,y_test=veri_olustur(test,time_step)
```

```python
x_train=x_train.reshape(x_train.shape[0],1,x_train.shape[1])

x_test=x_test.reshape(x_test.shape[0],1,x_test.shape[1])
```

```python
model=Sequential()

model.add(LSTM(128))
model.add(Dense(64))
model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")

model.fit(x_train,y_train,epochs=50)
```

```python
model.summary()
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred=scaler.inverse_transform(y_pred)
```

```python
y_test=scaler.inverse_transform(y_test)
```

```python
plt.figure(figsize=(8,5))
plt.plot(y_test,label="gerçek değer")
plt.plot(y_pred, label="tahmin değeri")

plt.xlabel("hafta")
plt.ylabel("bilet sayısı")

plt.legend()
plt.show()
```
