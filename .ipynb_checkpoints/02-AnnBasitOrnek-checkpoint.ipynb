```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv("datasets/CCPP/Train.csv")
test=pd.read_csv("datasets/CCPP/Test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
sample=pd.read_csv("datasets/CCPP/sample_submission.csv")
sample.head()
```

```python
test=pd.concat([test,sample],axis=1)
test.head()
```

```python
df=pd.concat([train,test],axis=0)
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=10,activation="relu",input_dim=4))
```

```python
ann.add(tf.keras.layers.Dense(units=8,activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(optimizer="adam",loss="mean_squared_error")
```

```python
ann.fit(x_train,y_train,epochs=20)
```

```python
y_pred=ann.predict(x_test)
```

```python
for i in y_test:
    print(i, end="   ")
```

```python
for i in y_pred:
    print(i, end="   ")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(r2_score(y_test,y_pred))
mean_squared_error(y_test,y_pred)
```
