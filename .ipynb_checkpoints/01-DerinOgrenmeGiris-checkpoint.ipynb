# Derin Öğrenme

Derin Öğrenme (Deep Learning), makine öğrenmesi yöntemlerinden biridir. Temel olarak, çok katmanlı yapay sinir ağlarının kullanıldığı bir makine öğrenmesi yaklaşımıdır. Derin Öğrenme, verilerdeki karmaşık örüntüleri ve ilişkileri otomatik olarak öğrenebilme yeteneğine sahiptir.

## Yapay Sinir Ağları

Yapay Sinir Ağları (Artificial Neural Networks), insan beynindeki nöronlar ve sinapslar arasındaki bağlantıları taklit eden bilgisayar sistemleridir. Temel olarak, yapay sinir ağları şu şekilde çalışır:

    Giriş Katmanı: Veriler, giriş katmanına aktarılır.

    Gizli Katmanlar: Giriş katmanından gelen veriler, çoklu gizli katmanlar aracılığıyla işlenir. Her gizli katman, verilerdeki örüntüleri ve ilişkileri öğrenmeye çalışır.

    Çıkış Katmanı: Son gizli katmandan gelen çıktılar, çıkış katmanında birleştirilir ve sonuç üretilir.

Yapay sinir ağlarının temel bileşenleri şunlardır:

    Nöronlar: Giriş verilerini alan ve işleyen temel birimlerdir.

    Ağırlıklar: Nöronlar arasındaki bağlantıların güçlerini temsil eder. Ağırlıklar, eğitim sırasında ayarlanır.

    Aktivasyon Fonksiyonları: Nöronların çıktılarını belirler. Sigmoid, ReLU, tanh gibi çeşitli aktivasyon fonksiyonları kullanılabilir.

    Eğitim Algoritmaları: Ağırlıkların optimum değerlere ulaşması için kullanılan algoritmalar, örneğin Geri Yayılım (Backpropagation) algoritması.

Yapay sinir ağları, makine öğrenmesi ve derin öğrenme alanlarında yaygın olarak kullanılır. Görüntü tanıma, doğal dil işleme, ses tanıma, robotik ve daha birçok alanda başarılı sonuçlar elde edilmektedir.

### Aktivasyon Fonksiyonları
Aktivasyon Fonksiyonları, yapay sinir ağlarındaki nöronların çıkış değerlerini belirleyen önemli bileşenlerdir. Aktivasyon fonksiyonları, nöronların toplam girdi değerini alarak, çıktı değerini hesaplar.

#### 1- Step Fonksiyonu
Bir eşik değeri alarak ikili bir sınıflandırma çıktısı (0 veya 1) üretir.

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    if x < 0:
        return 0
    else:
        return 1

x = np.arange(-5, 5, 0.1)
print(x[0:5])

y = [step_function(val) for val in x]

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2,label="step")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### 2- Sigmoid Fonksiyonu
En çok kullanılan fonksiyondur <br>
Sigmoid fonksiyonu, girdi değerlerini 0 ile 1 arasında bir değere dönüştürür. S-şeklindeki eğrisi nedeniyle, sınıflandırma problemlerinde yaygın olarak kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5, 5, 0.1)

y = sigmoid(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, label="sigmoid",color="C1")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sigmoid Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### 3- Tanh Fonksiyonu
Tanh fonksiyonu, girdi değerlerini -1 ile 1 arasında bir değere dönüştürür. Sigmoid fonksiyonuna benzer, ancak sıfır merkezlidir. Derin sinir ağlarında sıklıkla kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.arange(-5, 5, 0.1)

y = tanh(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, label="tanh",color="C5")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Tanh Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### 4- Relu Fonksiyonu
Gizli katmanlarda kullanılır <br>
ReLU fonksiyonu, girdi değeri pozitifse çıktı değerini girdi değerine eşitler, girdi değeri negatifse çıktı değerini 0 yapar. Derin sinir ağlarında yaygın olarak kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)

y = relu(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, label="ReLU",color="C2")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('ReLU Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### 5- Softmax
Softmax fonksiyonu, çok sınıflı sınıflandırma problemlerinde kullanılır. Girdi değerlerini olasılık değerlerine dönüştürür ve çıktı değerlerinin toplamı 1'e eşittir.

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

x = np.arange(-5, 5, 0.1)

y = softmax(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, color='red', label="Softmax")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Softmax Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### Softplus Fonksiyonu
Softplus fonksiyonu, girdi değerlerini 0'dan başlayarak yumuşak bir eğri oluşturacak şekilde dönüştürür. Regresyon problemlerinde kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def softplus(x):
    return np.log(1 + np.exp(x))

x = np.arange(-5, 5, 0.1)

y = softplus(x) 

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, color='C4', label="Softplus")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Softplus Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### ELU (Exponential Linear Unit) Fonksiyonu
ELU fonksiyonu, ReLU fonksiyonuna benzer, ancak negatif girdi değerleri için daha yumuşak bir geçiş sağlar. Görüntü işleme ve doğal dil işleme uygulamalarında kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-5, 5, 0.1)

y = elu(x) 

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, color='blue', label="ELU")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('ELU Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### PReLU (Parametric Rectified Linear Unit) Fonksiyonu
PReLU fonksiyonu, ReLU fonksiyonuna benzer, ancak negatif girdi değerleri için daha esnek bir eğim sağlar. Görüntü işleme ve doğal dil işleme uygulamalarında kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def prelu(x, alpha=0.1):
    return np.maximum(0, x) + alpha * np.minimum(0, x)

x = np.arange(-5, 5, 0.1)

y = prelu(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, color='purple', label="PReLU")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('PReLU Function')
plt.grid(True)

plt.legend()
plt.show()
```

#### Swish Fonksiyonu
Swish fonksiyonu, sigmoid fonksiyonuna benzer bir S-şeklindeki eğriye sahiptir. Girdi değeri arttıkça, çıktı değeri 1'e yaklaşır. Görüntü işleme ve doğal dil işleme uygulamalarında kullanılır.

```python
import numpy as np
import matplotlib.pyplot as plt

def swish(x, beta=1.0):
    return x * (1 / (1 + np.exp(-beta * x)))

x = np.arange(-5, 5, 0.1)

y = swish(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=2, color='orange', label="Swish")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Swish Function')
plt.grid(True)

plt.legend()
plt.show()
```
